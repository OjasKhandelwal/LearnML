```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
#plotting using matplotlib 
#scatter is used to scatter x , y points on 2D graph

plt.scatter(x , y , s=100, c='g' , marker = '*')  # s=sizeofmarkers     c=colour    marker=jis symbol se plot karna hai
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('graph in 2D')
plt.show()    #command to show graph
```

```python
#plot is used to connect points with line

plt.plot(x,y,linestyle='--',color='r',linewidth=2,marker='o',markersize=10 , label ='line1')


plt.grid(True, linestyle='--', alpha=0.7) #used to add grid to the plot (peeche jo lines dikh rahi hai)  
                               #alpha is used to control transparency b/w 0 to 1


plt.xlim(0, 6)  # Set x-axis limits
plt.ylim(0, 50)  # Set y-axis limits



#annotate is used to add text or arrow to highlight any specific point
plt.annotate('Max Value', xy=(5, 40), xytext=(3, 35),arrowprops=dict(facecolor='black', shrink=0.05)) 
              #text that has to be displayed
#xy: the point to annotate(x=5,y=40)  #xytext: position of the text ; as (x=3 ,y=35) , the text will be slightly away from xy
#arrowprops: Customizes the arrow 
#shrink:this slightly shrinks the arrow so it doesnt point directly at the exact (5, 40) location but is instead slightly shortened




plt.legend()  #we use legend only when we have added label to our plot elements
#legend is that small box on the top-left corner that labels the different elements (lines, bars, markers, etc.)


plt.show()
```

```python
#creating a bar graph
plt.bar(x, y, color='y', label='Bars')
plt.legend()
plt.show()
```

```python
#histogram - it show distribution of numerical data

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=2, color='purple', edgecolor='black') #bins -no. of bars
plt.show()
```

```python
#boxplot

data = [np.random.normal(0,std,100) for std in range(1,4)] #np.random.normal(): generates 100 random data points from a normal distribution
#0 is the mean of the distribution #std is the standard deviation #100 is the number of data points
#The code loops over the values std = 1, 2, 3 (range from 1 to 3).
#For each value of std, it generates a normal distribution with mean 0 and standard deviation std, and stores the resulting 100 values in the list data



plt.boxplot(data , vert=True , patch_artist=True)  #vert=True: makes the boxplots vertical #patch_artist=True: fills the boxes with color


plt.show()
```

```python
#piechart - shows proportion of categories

l = ['A','B','C','D'] #label of each pie
s = [15, 30, 45, 10]  #size of each pie
e = (0.4 , 0 ,0 ,0)  #exploding 1st pie (alag nikaalna)
plt.pie(s, explode = e,labels=l, autopct='%1.1f%%', startangle=90)  #autopct: Displays percentages on the chart    startangle: Rotates the pie chart.


plt.axis('equal')  # Ensures the pie is circular
plt.show()
```

```python
#plotting multiple plots at once (sub-plot)


#the subplot works as a array , so it has rows n columns & no. of plots = rXc 
#positions are also like array (0,1) so that means plot at 0th row n 1st colmn


fig, ax = plt.subplots(2, 2)  #    fig: the figure object, which contains all the subplots   
#ax: an array of axes , as we have 2 rows n 2 colmns , the array will be 2x2


ax[0, 0].plot(x, y)  # Plot in the first subplot
ax[0, 1].scatter(x, y)  # Plot in the second subplot
ax[1, 0].bar(x, y)  # Plot in the third subplot
ax[1, 1].hist(data, bins=4)  # Plot in the fourth subplot



plt.tight_layout()  # Adjust spacing b/w plots to ensure title n labels dont overlap



plt.show()
```

```python
#multiple lines on the same plot

y2 = [5, 15, 20, 25, 35]
plt.plot(x, y, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```
