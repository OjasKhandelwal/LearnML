```python
import torch
```

```python
x = torch.tensor(3.0 , requires_grad=True) #requries_grad ko true set karte hai to pytorch ko apne aap samajh aa jaata hai ki hum iss tensor ka derivative calculate karna chahte hai

y = x**2
```

```python
y.backward()   # isse dy/dx calculate hota hai
```

```python
x.grad  #dy/dx
```

```python
x.grad.zero_()   #kyuki gradients accumulate hote hai to last pass ka  gradient , current pass me add hota jaayega which is not desirable
                 #so we need to clear our gradients via zero_()
```

```python
#disable gradient tracking - training ke baad backward ki zarurat nahi hogi

#option 1 - requires_grad_(False)
#option 2 - detach()
#option 3 - torch.no_grad()


x.requires_grad_(False)
```

```python
z = x.detach()  #detach me hum ek new tensor banate hai , lekin uss new tensor me old tensor jesi exact values hongi lekin ye computational graph se alag ho chuka hai
```

```python
with torch.no_grad():
  y = x**2
```
